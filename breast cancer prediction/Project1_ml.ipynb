```python
#import libraries
import pandas as pn
import seaborn as sb
import opendatasets as op
```

```python
op.download("https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data")
```

```python
#load data on dataframe
df = pn.read_csv("breast-cancer-wisconsin-data\data.csv")
```

```python
#display dataframe
df.head()
```

```python
#count of rows and columns
df.shape
```

```python
#count number of empty values
df.isna().sum()
```

```python
#Drop the values of the column containing null values
df.dropna(axis=1,inplace=True)
```

```python
#count of rows and columns
df.shape
```

```python
#count of M and B diagnosis in our datasets
df['diagnosis'].value_counts()
```

```python
##label encoding
#get datatypes off each column in our dataset
df.dtypes
```

```python
#Encode the diagnosis values 
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1]=labelencoder.fit_transform(df.iloc[:,1].values)
```

```python
#display df
df
```

```python
#splitting data into dependent and independent datasets
X=df.iloc[:,2:].values
Y=df.iloc[:,1].values
```

```python
#splitting dataset into training 75% and testing 25%
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=.75,test_size=.25)
```

```python
#scaling the data(featuring the data)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#print data
print(X_train)
```

```python
#build a logistic regression classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
#make use if trained model to make predictions on test data
prediction = classifier.predict(X_test)
```

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,prediction)
print(cm)
sb.heatmap(cm,annot=True)
```

```python
#get accuracy score for model
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction))
```
